Extract the tgz file.\
Please see doc/ and examples/ in the package for more information.


```python
import glob
emcdir = 'EMC_interface/EMC'
has_emc = len(glob.glob(f'{emcdir}/v*/bin/emc*'))>0

if not has_emc:
    from subprocess import run
    tgzfile = glob.glob(f'{emcdir}/*tgz')[0]
    cmd = f'tar zxvf {tgzfile} -C {emcdir}'
    print(cmd)
    run(cmd, shell=True)
```
